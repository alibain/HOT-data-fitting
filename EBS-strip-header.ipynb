```python
import codecs
import os
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
# Import elastic scattering data - change encoding in order to read file and output data without header 
dname="C1Trace0" # name of file
f1=5942 # begining file number
f2=5952 # ending file number

i=f1
while i < f2+1:
    sourceFileName=os.path.join(r"C:\Users\zu21202\OneDrive - University of Bristol\Documents\UoB-PostDoc\HOT-data\raw",dname+str(i))
    targetFileName=r"C:\Users\zu21202\OneDrive - University of Bristol\Documents\UoB-PostDoc\HOT-data\fix.txt"
    outname=os.path.join(r'C:\Users\zu21202\OneDrive - University of Bristol\Documents\UoB-PostDoc\HOT-data\header-removed\df-'+dname+str(i))
    BLOCKSIZE = 1048576 # or some other, desired size in bytes
    with codecs.open(sourceFileName+'.txt', "r", "windows-1252") as sourceFile:
        with codecs.open(targetFileName, "w", "utf-8") as targetFile:
            while True:
                contents = sourceFile.read(BLOCKSIZE)
                if not contents:
                    break
                targetFile.write(contents)
    dat=pd.read_csv(targetFileName,delimiter='\t', skiprows=4)
    dat.to_csv(outname+'.txt')
    i=i+1
```

```python
k=0
i=f1-1 #Starting C1Trace file number - 1
while i< f2: # Ending C1Trance file nummber
    j=str(i+1)
    f=r'C:\Users\zu21202\OneDrive - University of Bristol\Documents\UoB-PostDoc\HOT-data\header-removed/df-C1Trace0'+j
    data = pd.read_csv(f+'.txt',index_col=0, sep=',')
    plt.plot(data.Time*1000**2,data.Ampl+k,linewidth=0.1)
    plt.xlim([-5,100])
    plt.xlabel('Time (micro seconds)')
    plt.ylabel('Amplitude (a.u.)')
    #plt.legend(power,bbox_to_anchor=(1,1), loc="upper left")
    i=i+1
    k=k+0.005
```
